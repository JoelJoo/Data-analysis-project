```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:/Data analysis freeCodeCamp/archive_3/Salary Prediction of Data Professions.csv")
df
```

```python
[df["SALARY"] >= 75000]
```

```python
specific_unit = ["IT", "Web", "Management", "Marketing"]
df[df["UNIT"].isin(specific_unit)]
```

```python

df[df["UNIT"].str.contains("Marketing")]
```

```python
df2_byJob = df.set_index("DESIGNATION")
df2_byJob
```

```python

df2_byJob.filter(items = ["SEX", "AGE"])
```

```python


df2_byJob.filter(like = "TOMASA" )
```
